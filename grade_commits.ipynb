```python
commits = {key:value for (key,value) in commits.items() if key in slugs}
```

```python
# make it a dataframe: (repo, auth)= [stuff]
df_dict={}

for team, l1 in commits.items():
    for auth, l2 in l1.items():
        if 'commits' not in l2.keys() or auth =='bad_committer':
            continue
        for sha, data in l2['commits'].items():
            idx_tuple = (team, auth, sha)
            row = data.values()
            df_dict[idx_tuple] = row
```

```python
mi = pd.MultiIndex.from_tuples(df_dict.keys(), names=['team', 'author', 'hash'])
commits_df=pd.DataFrame.from_dict(df_dict, orient='index', columns=['additions', 'commit_time', 'deletions','html_url', 'total', 'week'])
commits_df.index = mi
```

```python
completed_students = set()
grades_df = None
# read from completed
if grades_df_path.exists():
    grades_df = pd.from_csv(grades_df_path)
    completed_students = set(grades_df.index)
```

```python
# driver = driver.getcommits_df.iloc[0]['html_url']
# it'd be cool to display in editor but that's a rabbit hole
# I should add message
nocommit_authors = []
b=""
for team in slugs:
    team_name = team
    team = teams[team]
    members = [d for d in team['members'].values()]
    print(f"assesing team {team_name}")
    for author in members:
        if author['netid'] in completed_students:
            continue
        print("assessing author with netid {}".format(author['netid']))
        try:
            student_df = commits_df.xs(author['github'], level=1)
        except:
            print("No commits found for author")
            nocommit_authors.append(author)
            grades_dict[author['netid']] = [0]*len(rubric['Criteria'])
            continue

        # filter out insignificant
        student_df = student_df[student_df['additions'] > 2]
        student_df['html_url'].apply(webbrowser.open)

        display(student_df)

        # collect automated stuff
        automated_criteria = []
        automated_criteria.append(2 if (student_df['week'].nunique() >= 2) else student_df['week'].nunique())
        automated_criteria.append(4 if (student_df['additions'].count() >= 4) else student_df['additions'].count())
        automated_criteria.append(2 if (student_df[student_df['total'] > 100]['additions'].count() <= student_df['additions'].count()//2) else 0)

        subjective_criteria = []
        for criteria, out_of in zip(rubric['Criteria'][3:], rubric['Out of'][3:]):
            subjective_criteria.append(input(criteria + f": Out of {out_of}"))

        grades_dict[author['netid']] = automated_criteria + subjective_criteria
        if grades_df is None:
            grades_df = pd.DataFrame.from_dict(grades_dict, orient='index', columns=[*rubric['Criteria']])
            grades_df.to_csv(grades_df_path)
        else:
            grades_df = grades_df.append(pd.DataFrame.from_dict(grades_dict, orient='index', columns=[*rubric['Criteria']]))
            grades_df.to_csv(grades_df_path)
        b=input("Press enter to continue after complete")
        if(b!=""):
            # save progress to gradebook and quit 
            quit()
    
    if(b!=""): 
        3
        break
```
